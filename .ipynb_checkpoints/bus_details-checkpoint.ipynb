```python
# Importing libraries
from selenium import webdriver
from selenium.webdriver import ActionChains
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import TimeoutException, NoSuchElementException
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
import pandas as pd
```

```python
df_A= pd.read_csv("C:\\Users\\admin\\Desktop\\REDBUS_CSV\\andhra.csv")
```

```python
df_A
```

```python

# Function to navigate to a link and click elements matching the link
def navigation_and_click(driver, link):
    driver.get(link)
    time.sleep(2)
    elements = driver.find_elements(By.XPATH, f"//a[contains(@href,'{link}')]")
    for element in elements:
        element.click()
        time.sleep(2)  
        
# click elements to views bus

def click_button(driver):
    try:
        click_button = driver.find_element(By.XPATH, "//div[@class='button']")
        click_button.click()
    except:
        pass
    time.sleep(2) 

# Function to scroll the page until no further scrolling is possible

def scroll_page(driver):
    scrolling = True
    while scrolling:
        old_page_source = driver.page_source
        ActionChains(driver).send_keys(Keys.PAGE_DOWN).perform()
        time.sleep(10)
        new_page_source = driver.page_source
        if new_page_source == old_page_source:
            scrolling = False

# Function to extract bus details
def extracting_bus_details(driver):
    bus_name_elements = driver.find_elements(By.XPATH, "//div[@class='travels lh-24 f-bold d-color']")
    bus_type_elements = driver.find_elements(By.XPATH, "//div[@class='bus-type f-12 m-top-16 l-color evBus']")
    duration_elements = driver.find_elements(By.XPATH, "//*[@class='dur l-color lh-24']")
    try: 
        rating_elements = driver.find_elements(By.XPATH, "//div[@class='clearfix row-one']/div[@class='column-six p-right-10 w-10 fl']")
    except:
        rating_elements = []
    seat_elements = driver.find_elements(By.XPATH, "//div[contains(@class, 'seat-left')]")
    price_elements = driver.find_elements(By.XPATH, "//*[@class='fare d-block']")
    start_time_elements = driver.find_elements(By.XPATH, "//*[@class='dp-time f-19 d-color f-bold']")
    end_time_elements = driver.find_elements(By.XPATH, "//*[@class='bp-time f-19 d-color disp-Inline']")

    # Extracting text from each element and storing in lists
    bus_names = [bus.text for bus in bus_name_elements]
    bus_types = [type.text for type in bus_type_elements]
    durations = [total_time.text for total_time in duration_elements]
    ratings = [rate.text for rate in rating_elements]
    seats = [seat_available.text for seat_available in seat_elements]
    prices = [fare.text for fare in price_elements]
    start_times = [starttime.text for starttime in start_time_elements]
    end_times = [endtime.text for endtime in end_time_elements]
    
    return bus_names, bus_types, durations, ratings, seats, prices, start_times, end_times

# Initialize the WebDriver and maximize the window
driver = webdriver.Chrome()
driver.maximize_window()

# Initialize lists to store route and bus details
Route_names = []
Bus_names_A = []
Bus_types_A = []
Duration_A = []
Ratings_A = []
Seats_Available_A = []
Price_A = []
Departing_time_A = []
Reaching_Time_A = []
Route_links = []

# Loop through each row in the DataFrame to process routes
for index,row in df_A.iterrows():
    route=row["Route_name"]
    link=row["Route_link"]
    
    # Navigate to the link, click elements, scroll the page, and extract bus details
    navigation_and_click(driver, link)
    click_button(driver)
    scroll_page(driver)
    bus_details = extracting_bus_details(driver)
    bus_names, bus_types, durations, ratings, seats, prices, start_times, end_times = extracting_bus_details(driver)
    
    # Append the extracted details to the respective lists
    
    for bus_name in bus_names: 
        Bus_names_A.append(bus_name)
        Route_links.append(link)
        Route_names.append(route)
       
    for bus_type in bus_types:
        Bus_types_A.append(bus_type)
   
    for duration in durations:
        Duration_A.append(duration)
    
    for rating in ratings:
        Ratings_A.append(rating)
    
    for seat in seats:
        Seats_Available_A.append(seat)
    
    for price in prices:
        Price_A.append(price)
    
    for start_time in start_times:
        Departing_time_A.append(start_time)
    
    for end_time in end_times:
        Reaching_Time_A.append(end_time)
    

print("sucessful")
    
```

```python
data_A=pd.DataFrame({
    "Route_name":Route_names,
    "Bus_name":Bus_names_A,
    "Bus_type":Bus_types_A,
    "Duration":Duration_A,
    "Ratings":Ratings_A,
    "Seats_Available":Seats_Available_A, 
    "Price":Price_A,
    "Departing_time":Departing_time_A, 
    "Reaching_Time":Reaching_Time_A,
    "Route_link":Route_links
})
```

```python
# Initialize the WebDriver
driver_WB = webdriver.Chrome()
driver_WB.maximize_window()

# Initialize lists to store route and bus details 
Route_names = []
Bus_names_WB = []
Bus_types_WB = []
Duration_WB = []
Ratings_WB = []
Seats_Available_WB = []
Price_WB = []
Departing_time_WB = []
Reaching_Time_WB = []
Route_links = []

# Loop through each row in the DataFrame to process routes
for index,row in df_WB.iterrows():
    route=row["Route_name"]
    link=row["Route_link"]
    
    # Navigate to the link, click elements, scroll the page, and extract bus details
    navigation_and_click(driver_WB, link)
    click_button(driver_WB)
    scroll_page(driver_WB)
    bus_details = extracting_bus_details(driver_WB)
    bus_names, bus_types, durations, ratings, seats, prices, start_times, end_times = extracting_bus_details(driver_WB)
    
    # Append the extracted details to the respective lists
    for bus_name in bus_names: 
        Bus_names_WB.append(bus_name)
        Route_links.append(link)
        Route_names.append(route)
       
    for bus_type in bus_types:
        Bus_types_WB.append(bus_type)
   
    for duration in durations:
        Duration_WB.append(duration)
    
    for rating in ratings:
        Ratings_WB.append(rating)
    
    for seat in seats:
        Seats_Available_WB.append(seat)
    
    for price in prices:
        Price_WB.append(price)
    
    for start_time in start_times:
        Departing_time_WB.append(start_time)
    
    for end_time in end_times:
        Reaching_Time_WB.append(end_time)
    

print("sucessful")
```

```python
data_WB=pd.DataFrame({
    "Route_name":Route_names,
    "Bus_name":Bus_names_WB,
    "Bus_type":Bus_types_WB,
    "Duration":Duration_WB,
    "Ratings":Ratings_WB,
    "Seats_Available":Seats_Available_WB, 
    "Price":Price_WB,
    "Departing_time":Departing_time_WB, 
    "Reaching_Time":Reaching_Time_WB,
    "Route_link":Route_links
})
```
